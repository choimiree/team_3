```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
def scroll(driver, timeout):
    scroll_pause_time = timeout

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(scroll_pause_time)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            # If heights are the same it will exit the function
            break
        last_height = new_height
```

```python
nickname = []
rating = []
title=[]
voice=[]
author=[]

for i in range(len(soup.select('ul.audiobook-list-component__list>li.audiobook-list-item-component>h3.title'))):
    driver.find_element_by_xpath(f'//*[@id="audiobookIndex"]/div[2]/div/ul/li[{i+1}]').click() #특정 책으로 들어가
    scroll(driver, 1) #스크롤 끝까지 내려
    html = driver.page_source
    soup2 = BeautifulSoup(html, 'html.parser') #특정 책 soup
    n = int(soup2.select('div.audiobook-reviews-component>h3.title>span')[0].text.split(' ')[1].strip('명')) #n정의
    for i in range((n-3)//12+1):
        더보기 = driver.find_element_by_xpath('//*[@id="audiobookDetails"]/div[5]/div/button[1]')
        더보기.click()
    time.sleep(1)
    html = driver.page_source
    soup3 = BeautifulSoup(html, 'html.parser') #댓글 다 나와있는 soup
    for i in range(n):
        c=soup3.select('dt.title')[0].text
        a=soup3.select('h3.nickname')[i].text
        b=soup3.select('div.star_rating>span')[i].get('style').split(' ')[1][:-2]
        d=soup3.select('dl.audiobook-details__info>dd')[1].text[3:]
        e=soup3.select('dl.audiobook-details__info>dd')[2].text[2:]
        nickname.append(a)
        rating.append(b)
        title.append(c)
        voice.append(d)
        author.append(e)
    driver.back() # 책 목록으로 돌아가
    scroll(driver, 1) #스크롤 끝까지 내려
```

# 책 두 권 이상에 댓글을 단 사람들

```python
word_cnt = dict() 

for name in nickname: 
    if name not in word_cnt.keys(): 
        word_cnt[name] = 1 
    else: 
        word_cnt[name] += 1
print(word_cnt)
```

```python
len(word_cnt)
```

```python
over_2_dict = {}
for key in word_cnt.keys() :
    if word_cnt.get(key) >= 2 :
        over_2_dict[key] = word_cnt.get(key)
```
